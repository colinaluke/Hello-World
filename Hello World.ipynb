```python
import numpy as np
import matplotlib.pyplot as plt
import datetime

# Generate test data using numpy
num_points = 100
x = np.linspace(0, 10, num_points) 
y = np.sin(x) 

# Create a time series graph
plt.figure(figsize=(10, 6)) 
plt.plot(x, y, label='Data', color='blue', linestyle='-')
plt.xlabel('Seconds')
plt.ylabel('Units')
plt.title('Hello World Plot')
plt.grid(True)
plt.legend()
plt.show()
```
